```python
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl_df = pd.read_csv('C:\\Users\\mozam\\Downloads\\data\\archive\\EPL_20_21.csv')
epl_df.head()
```

```python
epl_df.info()
```

```python
epl_df.describe()
```

```python
#create column

epl_df['MinsPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)
epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)
epl_df
```

```python
#Total Goals

Total_Goals = epl_df['Goals'].sum()
print("Total number of goals in 2020-21 season:",Total_Goals)
```

```python
#Total Penalty Goals

Total_Penalty_Goals = epl_df['Penalty_Goals'].sum()
print("Total number of penalty goals in 2020-21 season:",Total_Penalty_Goals)
```

```python
#Total Penalty Attempts

Total_Penalty_Attempts = epl_df['Penalty_Attempted'].sum()
print("Total number of penalty Attempts in 2020-21 season:",Total_Penalty_Attempts)
```

```python
# Pie chart for penalties missed vs scored

plt.figure(figsize=(13, 6))
pl_not_scored = Total_Penalty_Attempts - Total_Penalty_Goals
data = [pl_not_scored, Total_Penalty_Goals]
labels = ['Penalties Missed', 'Penalties Scored']
color = sns.color_palette('crest')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%')
plt.show()
```

```python
# most player from which countries

epl_df.groupby('Nationality').size().sort_values(ascending = False)
```

```python
#most player of country wise bar chart

nationality = epl_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize = (12,6), color = sns.color_palette("magma"))
```

```python
#club with maximum players in their squad

epl_df['Club'].value_counts()
```

```python
#club with maximum players in their squad

epl_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette("viridis"))
```

```python
#club with least players in their squad

epl_df['Club'].value_counts().nsmallest(5).plot(kind='bar', color=sns.color_palette("viridis"))
```

```python
# players based on age group

Under20 = epl_df[epl_df['Age'] <= 20]
Age20_25 = epl_df[(epl_df['Age'] > 20) & (epl_df['Age'] <= 25)]
Age25_30 = epl_df[(epl_df['Age'] > 25) & (epl_df['Age'] <= 30)]
Above30 = epl_df[epl_df['Age'] > 30]
```

```python
x = np.array([Under20['Name'].count(), Age20_25['Name'].count(), Age25_30['Name'].count(), Above30['Name'].count()])
mylabels = ["Less than 20", "Between 20 and 25", "Between 25 and 30", "Above 30"]
plt.title('Total player with age group',fontsize = 20)
plt.pie(x,labels = mylabels, autopct="%.1f%%")
plt.show()
```

```python
# Total under 20 players in each club

players_under_20 = epl_df[epl_df['Age'] < 20]
players_under_20['Club'].value_counts().plot(kind='bar', color=sns.color_palette("cubehelix"))
```

```python
#Under 20 players in ManU

players_under_20[players_under_20['Club']=='Manchester United']
```

```python
#Under 20 players in Chelsea

players_under_20[players_under_20['Club']=='Chelsea']
```

```python
# Average age of players in each club

plt.figure(figsize=(12,6))
sns.boxplot(x = 'Club', y = 'Age', data = epl_df)
plt.xticks(rotation = 90)
```

```python
num_player = epl_df.groupby('Club').size()
data = (epl_df.groupby('Club')['Age'].sum())/num_player
data.sort_values(ascending = False)
```

```python
# Total assists from each club

Assists_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Assists'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y = 'Assists', data = Assists_by_clubs.sort_values(by='Assists', ascending=False),palette='Set2')
ax.set_xlabel('Club',fontsize=30)
ax.set_ylabel('Assists', fontsize=20)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = (20,8)
plt.title('Plot of clubs vs total assists',fontsize=20)
```

```python
# top 10 assists

top_10_assists = epl_df[['Name', 'Club', 'Assists', 'Matches']].nlargest(n = 10, columns = 'Assists')
top_10_assists
```

```python
#total goals by club

Goals_by_clubs = pd.DataFrame(epl_df.groupby('Club', as_index = False)['Goals'].sum())
sns.set_theme(style='whitegrid', color_codes=True)
ax = sns.barplot(x='Club', y = 'Goals', data = Goals_by_clubs.sort_values(by='Goals', ascending=False),palette='rocket')
ax.set_xlabel('Club',fontsize=40)
ax.set_ylabel('Goals', fontsize=40)
plt.xticks(rotation=90)
plt.rcParams['figure.figsize'] = (15,8)
plt.title('Plot of clubs vs total goals',fontsize=20)
```
